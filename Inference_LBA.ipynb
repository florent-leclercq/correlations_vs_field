```python
# Summary statistics: infer alpha and beta.  Use Sellentin-Heavens likelihood.

def lnL_LN(LN,alpha,beta,xi,invcov,nsims,PixelNoise):
    """
    Compute the ln likelihood
    
    Parameters:
        LN : LogNormalField instance
            log-normal field setup
        alpha : float
            nongaussianity parameter in lognormal model
        beta  : float
            width of gaussian correlation function
        xi : float array
            two-point correlation function (the data)
        invcov : float matrix
            Inverse covariance matrix of the data (ndata x ndata)
        nsims : int
            Number of simulations used to estimate the covariance matrix
        PixelNoise : float
            Standard deviation of added noise per pixel
    """
    
    ave_r   = LN.average_r
    ndata   = LN.nonzerobins
    ndatax  = xi.size
    ndatac1 = invcov.shape[0]
    ndatac2 = invcov.shape[1]
    
    assert ndata == ndatax
    assert ndata == ndatac1
    assert ndata == ndatac2
    
    if(alpha<0):
        return -np.inf
    
    data = xi
    invC = invcov
    
    theory = LN.xi_LN(ave_r,beta,alpha,PixelNoise)
    
    lnLike_SH   = Sellentin_Heavens(data,theory,invC,nsims)
    
    return lnLike_SH

def Sellentin_Heavens(data,theory,invC,nsims):
    """
    Compute the Sellentin-Heavens likelihood (Sellentin & Heavens 2016)
    
    Parameters:
        data : float (array of size N)
            data vector (xi, in this case)
        theory : float (array of size N)
            theory vector
        invC : float (N x N)
            inverse covariance matrix
        nsims : int
            Number of simulations used to estimate the covariance matrix        
    """
        
    diff   = data-theory
    chisq  = np.dot(diff,np.dot(invC,diff))
    
    lnL_SH =  -0.5*nsims*np.log(1+chisq/(nsims-1)) 
    
    return lnL_SH
```

```python
PixelNoises = [0.01, 0.1, 1]

try:
    with open(dataout+'_dependence_noise_sims.pkl', 'rb') as fp:
        lognormfields = pickle.load(fp)
except IOError:
    lognormfields = {}
    for PixelNoise in PixelNoises:
        print("PixelNoise =", PixelNoise)

        LN=LogNormalField(Lside,rmax,nbin)

        # Create a list of lognormal field objects and compute their correlation functions:
        lognormfields[PixelNoise] = []

        for i in range(1,nsims+1):

            if(i % 100==0 ):
                print('Simulation',i,'/',nsims,' done')

            sim = LN.run_simulation(alpha,beta,PixelNoise)
            cor = LN.compute_corrfn(sim)

            lognormfields[PixelNoise].append(np.array((sim,cor),dtype=LN.dt))
    
    with open(dataout+'_dependence_noise_sims.pkl', 'wb') as fp:
        pickle.dump(lognormfields, fp, protocol=pickle.HIGHEST_PROTOCOL)

try:
    with open(dataout+'_dependence_noise.pkl', 'rb') as fp:
        lnLgt_PixelNoise = pickle.load(fp)
except IOError:
    lnLgt_PixelNoise = {}
    for PixelNoise in PixelNoises:
        print("PixelNoise =", PixelNoise)

        LN=LogNormalField(Lside,rmax,nbin)

        # Compute the mean correlation function from the list of LN field objects, and its covariance
        mean2pt, cov = LN.covariance(lognormfields[PixelNoise])

        print('Condition number of covariance matrix =',np.linalg.cond(cov))
        invcov   = np.linalg.inv(cov)

        # Get the likelihood of the ground truth
        for i in range(1,nsims+1):

            if(i % 100==0 ):
                print('Likelihood',i,'/',nsims,' done')

            from scipy.optimize import minimize
            def target(x):
                data = lognormfields[PixelNoise][i-1]['corrfn']
                return -lnL_LN(LN, x[0], x[1], data, invcov, nsims, PixelNoise)

            with warnings.catch_warnings():
                warnings.simplefilter("ignore")
                res = minimize(target, np.array([alpha,beta]),
                           options={'disp': False})
            C = -target(res['x'])

            lnLgt_PixelNoise[PixelNoise,i-1] = lnL_LN(LN,alpha,beta,lognormfields[PixelNoise][i-1]['corrfn'],invcov,nsims,PixelNoise) - C

    with open(dataout+'_dependence_noise.pkl', 'wb') as fp:
        pickle.dump(lnLgt_PixelNoise, fp, protocol=pickle.HIGHEST_PROTOCOL)
```
